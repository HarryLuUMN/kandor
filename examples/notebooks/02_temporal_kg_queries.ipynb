# Temporal KG Queries

This notebook focuses on memory retrieval APIs:

- `get_world_state_at`
- `get_entity_history`
- `get_active_relations`
- `get_events_between`
- `get_relevant_context`


```python
from kandor.core.models import Event, Relation
from kandor.memory.temporal_kg import TemporalKGMemory
```

```python
memory = TemporalKGMemory()

memory.add_relation(
    Relation(
        subject="city.alpha",
        predicate="controlled_by",
        object="kingdom.red",
        start_time=0,
        end_time=2,
    )
)

memory.add_relation(
    Relation(
        subject="city.alpha",
        predicate="controlled_by",
        object="kingdom.blue",
        start_time=3,
    )
)

memory.add_relation(
    Relation(
        subject="city.alpha",
        predicate="trades_with",
        object="city.beta",
        start_time=1,
    )
)

memory.add_event(
    Event(
        id="event.siege",
        type="battle",
        timestamp=2,
        participants=["city.alpha", "kingdom.red", "kingdom.blue"],
        summary="A siege weakens the old rulers of Alpha.",
    )
)

memory.add_event(
    Event(
        id="event.transfer",
        type="ruler_changed",
        timestamp=3,
        participants=["city.alpha", "kingdom.blue"],
        summary="Kingdom Blue takes control of Alpha.",
    )
)
```

```python
memory.get_world_state_at(2)
```

```python
memory.get_world_state_at(3)
```

```python
memory.get_entity_history("city.alpha")
```

```python
memory.get_events_between(2, 3)
```

```python
memory.get_relevant_context(["city.alpha"], time=3, horizon=1)
```
