# Temporal KG Widget

This notebook shows the notebook widget for exploring a saved temporal KG snapshot.

Before running it, install widget extras in the kernel you plan to use:

```bash
python3 -m pip install 'kandor[widgets]'
```


````python
from pathlib import Path
import sys

for candidate in (Path.cwd(), Path.cwd().parent, Path.cwd().parent.parent):
    if (candidate / 'kandor').exists():
        sys.path.insert(0, str(candidate))
        break
````

````python
from kandor import TemporalKGWidget, load_temporal_kg_snapshot
````

````python
from pathlib import Path

for candidate in (
    Path('reports/smoke/openai-smoke-kg-2026-03-24.json'),
    Path('../reports/smoke/openai-smoke-kg-2026-03-24.json'),
    Path('../../reports/smoke/openai-smoke-kg-2026-03-24.json'),
):
    if candidate.exists():
        snapshot_path = candidate
        break
else:
    raise FileNotFoundError('Could not locate the bundled smoke KG snapshot.')

data = load_temporal_kg_snapshot(snapshot_path)
data.keys()
````

````python
widget = TemporalKGWidget(snapshot_path)
widget
````

If you want to start from in-memory data instead of a file, pass the parsed dictionary directly:

```python
widget = TemporalKGWidget(data)
widget
```
